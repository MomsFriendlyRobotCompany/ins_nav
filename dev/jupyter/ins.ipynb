```python
import numpy as np
from the_collector import BagIt, Pickle
from matplotlib import pyplot as plt
```

```python
def info(bag):
    print('Bag keys:')
    print('-'*50)
    for k in bag.keys():
        print(f'  {k:>10}: {len(bag[k]):<7}')
        
def plot3(t, data,title):
    x = [x[0] for x in data]
    y = [x[1] for x in data]
    z = [x[2] for x in data]
    plt.plot(t,x,label='x')
    plt.plot(t,y,label='y')
    plt.plot(t,z,label='z')
    plt.title(title)
    plt.legend()
    plt.grid(True);
```

```python
bagfile = "../path-data/test-simple.pickle.bag"
bag = BagIt(Pickle)
data = bag.read(bagfile)
info(data)
```

```python
accel = np.array([x[0] for x in data["imu"]])
gyro = np.array([x[1] for x in data["imu"]])
mag = np.array([x[2] for x in data["imu"]])

start = data["imu"][0][3]
t = np.array([x[3]-start for x in data["imu"]])
```

```python
plot3(t,accel,"Accel[g]")
```

```python
plot3(t,gyro,"Gyro[rad/s]")
```

```python
plot3(t,mag,"Mag[uT]")
```
